```python
import matplotlib.pyplot as plt 

name = 'Stepa'
```

```python
with open(f'data_{name}.txt', encoding='UTF-8') as file_in:
    lines = file_in.readlines()
    results = [line.split() for line in lines]

    without_parallel = [float(x) for x in results[0]]
    openMP_parallel = [float(x) for x in results[1]]
    manual_parallel = [float(x) for x in results[2]]

matrix_sizes = list(range(50, 3050, 50))
```

```python
fig, ax = plt.subplots(figsize=(14, 10))    
plt.plot(matrix_sizes, manual_parallel) 
#plt.title("Без распараллеливания - вариант 3")
#plt.title("С использованием OpenMP - вариант 3")
plt.title("Своя реализация распараллеливания - вариант 3")
plt.xlabel("Размер матрицы")
plt.ylabel("Время выполнения (секунды)")
plt.grid()
plt.show()  
```
