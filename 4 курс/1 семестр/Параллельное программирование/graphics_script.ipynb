```python
import matplotlib.pyplot as plt 

name = 'Dima'
```

```python
with open(f'data_{name}.txt', encoding='UTF-8') as file_in:
    lines = file_in.readlines()
    results = [line.split() for line in lines]

    without_parallel = [float(x) for x in results[0]]
    openMP_parallel = [float(x) for x in results[1]]
    manual_parallel = [float(x) for x in results[2]]

matrix_sizes = list(range(50, 3050, 50))
```

```python
a1 = [x / y for x, y in zip(without_parallel[1:], openMP_parallel[1:])]
a2 = [x / y for x, y in zip(without_parallel[1:], manual_parallel[1:])]
```

```python
fig, ax = plt.subplots(figsize=(10, 6))    
plt.plot(matrix_sizes[1:], a1, color='blue', label='Ускорение OpenMP')
plt.plot(matrix_sizes[1:], a2, color='red', label='Ускорение своей реализацией')
# plt.plot(matrix_sizes, manual_parallel, color='red', label='Своя реализация распараллеливания')
plt.title("Графики Колбанов")
plt.xlabel("Размер матрицы")
plt.ylabel("Ускорение")
plt.legend()
plt.grid()
plt.show()  
```
